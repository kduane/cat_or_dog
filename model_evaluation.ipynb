```python
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.models import load_model
```

```python
#load cnn model
cnn = load_model('model')
```

```python
sample_path = 'data/cat/Russian Blue/41156128_7499.jpg'
def prep(filepath):
    size = (160, 160)
    img_arr = cv2.imread(sample_path)
    #convert to color array
    img_rgb = cv2.cvtColor(img_arr, cv2.COLOR_BGR2RGB)
    #reshape to uniform size 160x160x3
    new_arr = cv2.resize(img_rgb, size)
    #set batch size of 1
    new_arr = new_arr.reshape(-1, 160, 160, 3)
    #normalize to match model preprocessing
    result = new_arr / 255
    return result
```

```python
plt.imshow(cv2.imread(sample_path))
```

```python
sample = prep(sample_path)
cnn.predict(sample)
```
